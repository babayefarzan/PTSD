```python
df = df.merge(df_pcl1, on="ID")
df = df.merge(df_pcl2, suffixes=('_pcl1', '_pcl2'), on="ID")
df = df.merge(df_pcl3.drop(['PCL3_Strict','pcl3', 'PCL3_Broad'],axis=1), on="ID")
```

```python
df.drop(ID, inplace=True, axis=1)
df.dropna(thresh=0.5, inplace=True, axis = 0)
df.dropna(thresh=0.5, inplace=True, axis = 1)
```

```python
mi = KNN()
df = pd.DataFrame(mi.fit_transform(df), columns=df.columns)
```

```python
X = df[features]
Y = df[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.1, random_state=271828, stratify=Y)
pipe = Pipeline(steps=[
            ('SMOTE', SMOTE(random_state=27)),
       ('RFE',  RFE(RandomForestClassifier(n_estimators=100), 30)),
                ('classifier', RandomForestClassifier())])
```

```python

params_grid = [
           {
            'classifier__n_estimators':[10,30,50,70,90,150,300],
            'SMOTE__k_neighbors':[4, 2, 3],
            'SMOTE__sampling_strategy':[0.8],
        #    'RFE__n_features_to_select':[30]
            }]
```

```python
gs = GridSearchCV(pipe, params_grid, cv=5, scoring='precision')
```

```python
gs.fit(X_train, y_train)
gs.best_score_
```
